```python
import numpy as np 
def save_numpy_array(data,filename=""):
    path = '/Users/paul.tgr/aPavel/VAE_diploma/diploma_scripts/filters/saved_metrics/'
    np.savetxt(path + filename+ '.csv', data, delimiter=',')

def load_numpy_array(filename=""):
    path = '/Users/paul.tgr/aPavel/VAE_diploma/diploma_scripts/filters/saved_metrics/'
    return np.loadtxt(path + filename+'.csv', delimiter=',')
```

```python
def save_metrics(arr_vars,arr_names):
    for i in range(len(arr_vars)):
        save_numpy_array(arr_vars[i],arr_names[i])

def load_metrics(arr_names):
    var = []
    for i in range(len(arr_names)):
        var.append(load_numpy_array(arr_names[i]))
    return var
```

```python
save_metrics(MSE_vars,MSE_names)
save_metrics(SSIM_vars,SSIM_names)
save_metrics(NCC_vars,NCC_names)
save_metrics(FS_SCORE_vars,FS_SCORE_names)
```
